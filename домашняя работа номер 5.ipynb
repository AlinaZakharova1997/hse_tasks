```python
import re
from collections import Counter, defaultdict


def get_chars_list(text: str) -> dict:
    """Step 1. Get chars from an input text"""
    chars = list(text)
    return chars

def get_pairs_with_frequency(chars: list) -> dict:
    """Step 2. Get counts of pairs of consecutive symbols"""
    pairs = {}
    #print(len(chars),'len chars') # 100000
    for i, char in enumerate(chars[:-1]):
        # делаем пары и добавляем их в словарь
        pair = (chars[i], chars[i + 1])
        if pair not in pairs:
            pairs.setdefault(pair,0)
        else:
            pairs[pair] += 1
    pairs = {k: pairs[k] for k in sorted(pairs, key=pairs.get, reverse=True)}
    #print(len(pairs))
    #print(pairs)
    return pairs

def get_words(pairs:dict, chars:list, k:int) -> list:
    # отсортированные по частоте ключи теперь лежат в списке
    keys = list(pairs.keys())
    top = []
    #print(chars)
    # делаем топ к пар
    for i, key in enumerate(keys):
        if i < k:
            top.append(keys[i])
    #print(top)
    words = []
    # идем по всему списку пар
    word_count = 0
    for i, char in enumerate(chars[:-1]):
        # если пара есть в топе
        if (chars[i], chars[i+1]) in top and word_count < k:
            # приплюсовываем ее к результирующему слову
            words.append(chars[i] + chars[i+1])
        else:
            words.append(chars[i])
        word_count+=1
    #print(words)
    return words


text = open('Рабочий стол/lenta.txt').read()[:100000]
#vocab = get_chars_list(text)  # Step 1
#pairs = get_pairs_with_frequency(vocab)
num_merges = 100  # Hyperparameter, в задании это N 
for i in range(num_merges):
    vocab = get_chars_list(text) 
    pairs = get_pairs_with_frequency(vocab)  # Step 2

    if not pairs:
        break
    # step 3
    vocab = get_chars_list(text) 
    vocabular = get_words(pairs, vocab, 100) 
#print(vocabular)    
# этот словарь надо зафиксировать
vocab_copy = vocabular.copy()
news = open('Рабочий стол/news.txt').read()

#print(best)
#print(vocab)
```
